```python
import json
india_states=json.load(open("states_india.geojson","r"))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data_Frame=pd.read_csv('lane-accidents.csv')
print(Data_Frame)
```

```python
state_id_map={}
```

```python
for feature in india_states['features']:
    feature['id']=feature['properties']['state_code']
    state_id_map[feature['properties']['st_nm']]= feature['id']
```

```python
state_id_map
```

```python
Data_Frame
```

```python
Data_Frame.at[33,'State/UT']='NCT of Delhi'
```

```python
Data_Frame.at[1,'State/UT']='Arunanchal Pradesh'
```

```python
Data_Frame.at[19,'State/UT']='Odisha'
```

```python
Data_Frame.at[29,'State/UT']='Andaman & Nicobar Island'
```

```python
Data_Frame.at[31,'State/UT']='Dadara & Nagar Havelli'
```

```python
Data_Frame=Data_Frame.drop(36)
```

```python
Data_Frame
```

```python
Data_Frame['id']=Data_Frame['State/UT'].apply(lambda x: state_id_map[x])
```

```python
import plotly.express as px
```

```python
import plotly.io as pio
pio.renderers.default = 'browser'
```

```python
fig=px.choropleth(data_frame=Data_Frame,locations='id',geojson=india_states,color='Single Lane - Accident - 2014 per 1L people',scope='asia')
fig.update_geos(fitbounds="locations", visible=False)
fig.show()

fig=px.choropleth(data_frame=Data_Frame,locations='id',geojson=india_states,color='Two Lanes - Accident - 2014 per 1L people',scope='asia')
fig.update_geos(fitbounds="locations", visible=False)
fig.show()

fig=px.choropleth(data_frame=Data_Frame,locations='id',geojson=india_states,color='3 Lanes or more w.o Median - Accident - 2014 per 1L people',scope='asia')
fig.update_geos(fitbounds="locations", visible=False)
fig.show()

fig=px.choropleth(data_frame=Data_Frame,locations='id',geojson=india_states,color='4 Lanes with Median - Accident - 2014 per 1L people',scope='asia')
fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
