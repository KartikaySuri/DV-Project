```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash_html_components import Label
from pandas.io.formats import style
import plotly.express as px
import pandas as pd
from dash.dependencies import Input, Output
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
india_states=json.load(open("states_india.geojson","r"))
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'browser'
```

```python
app = dash.Dash(
    __name__,
)

Status=['Accident per 1L population','Injury per 1L population','Death per 1L population']
Lane=['Single Lane','Two Lane','Three Lane','Four Lane']

colors = {"background": "#0091D5", "text": "white"}

main_style = {
    'background': '#FFFFFF',
    'font-family': "Arial"
}

title_style = {
    'background': '#5473FF',
    'text': '#FFFFFF'
}

desc_style = {
    'background': '#FFFFFF',
    'text': '5473FF'
}
dropdown_style = {
    'background': '#DDE5ED',
    'text': '#221C35'
}

app.layout = html.Div( style={
        'backgroundColor': '#0091D5',
        'font-family': main_style['font-family']
    },children=
    [
        html.H1(children="Road Infrastructure and Impact on Casualty",
        style={
            'backgroundColor': '#1c4e80',
            'color': title_style['text'],
            'textAlign': 'center'}   
        ),
        html.Div(children='This dashboard allows user to see the impact of Road Infrastructre on Casualty across the Country differentiated by State',
        style={
            'backgroundColor': '#4cb5f5',
            'color': desc_style['text'],
            'textAlign': 'center'}),
        html.Div(
            [
                html.Div( style={
            'backgroundColor': 'cyan',
            'color': 'black',
            'textAlign': 'left',"width": "50%"},children=
                    [
                        html.Label("Status"),
                        dcc.Dropdown(
                            id="status-dropdown",style={'color':'#1c4e80','backgroundcolor':'cyan'},
                            options=[
                                {"label": s, "value": s} for s in Status
                            ],
                            className="dropdown"
                        ),
                        html.Label("Lane Wise Statistics"),
                        dcc.Dropdown(
                            id="lane-dropdown",style={'color':'#1c4e80','backgroundcolor':'cyan'},
                            options=[
                                {"label": y, "value": y}
                                for y in Lane
                            ],
                            className="dropdown",
                        )
                    ]
                ),
                
            ],
            className="row",
        ),
        html.Div(dcc.Graph(id="life-exp-vs-road-infrastructure",style={'width': '200vh', 'height': '80vh'}), className="chart"),
        
    ],
    className="container")


@app.callback(
    Output("life-exp-vs-road-infrastructure", "figure"),
    Input("status-dropdown", "value"),
    Input("lane-dropdown", "value"),
)
def update_figure(status, lane_status):
    if status == 'Accident per 1L population':
        Data_Frame=pd.read_csv('accidents.csv')
    elif status == 'Injury per 1L population':
        Data_Frame=pd.read_csv('injured.csv')
    elif status == 'Death per 1L population':
        Data_Frame=pd.read_csv('killed.csv')

    state_id_map={}
    for feature in india_states['features']:
        feature['id']=feature['properties']['state_code']
        state_id_map[feature['properties']['st_nm']]= feature['id']
    Data_Frame.at[33,'State/UT']='NCT of Delhi'
    Data_Frame.at[1,'State/UT']='Arunanchal Pradesh'
    Data_Frame.at[19,'State/UT']='Odisha'
    Data_Frame.at[29,'State/UT']='Andaman & Nicobar Island'
    Data_Frame.at[31,'State/UT']='Dadara & Nagar Havelli'
    Data_Frame=Data_Frame.drop(36)
    Data_Frame['id']=Data_Frame['State/UT'].apply(lambda x: state_id_map[x])



    fig = px.choropleth(data_frame=Data_Frame,locations='id',hover_name=Data_Frame['State/UT'],
    geojson=india_states,color=lane_status,scope='asia').update_layout(
        plot_bgcolor=colors["background"],
        paper_bgcolor=colors["background"],
        font_color=colors["text"],
    )

    return fig
if __name__ == "__main__":
    app.run_server(debug=False)
```
